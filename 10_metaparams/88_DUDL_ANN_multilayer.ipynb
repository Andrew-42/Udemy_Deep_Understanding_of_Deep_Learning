```python
# import libraries
import torch
import torch.nn as nn
import matplotlib.pyplot as plt
import numpy as np
```

```python
# create data

nPerClust = 100
blur = 1

A = [  -2,  1 ]
B = [  3, 1 ]

# generate data
a = [ A[0]+np.random.randn(nPerClust)*blur , A[1]+np.random.randn(nPerClust)*blur ]
b = [ B[0]+np.random.randn(nPerClust)*blur , B[1]+np.random.randn(nPerClust)*blur ]

# true labels
labels_np = np.vstack((np.zeros((nPerClust,1)),np.ones((nPerClust,1))))

# concatanate into a matrix
data_np = np.hstack((a,b)).T

# convert to a pytorch tensor
data = torch.tensor(data_np).float()
labels = torch.tensor(labels_np).float()

# show the data
fig = plt.figure(figsize=(5,5))
plt.plot(data[np.where(labels==0)[0],0],data[np.where(labels==0)[0],1],'bs')
plt.plot(data[np.where(labels==1)[0],0],data[np.where(labels==1)[0],1],'ko')
plt.title('The qwerties!')
plt.xlabel('qwerty dimension 1')
plt.ylabel('qwerty dimension 2')
plt.show()
```

```python
from torch.utils.data import DataLoader,TensorDataset

train_dataDataset = TensorDataset(data, labels)

train_loader = DataLoader(train_dataDataset, batch_size=32, shuffle=True, drop_last=True)
```

# Functions to build and train the model

```python
import torch.nn.functional as F


class Model(nn.Module):
  def __init__(self):
    super().__init__()

    ### input layer
    self.input = nn.Linear(2,32)
    
    ### hidden layers
    self.bnorm1 = nn.BatchNorm1d(32) # the number of units into this layer
    self.fc1    = nn.Linear(32,32)
    self.bnorm2 = nn.BatchNorm1d(32) # the number of units into this layer
    self.fc2    = nn.Linear(32,1)

    ### output layer
    self.output = nn.Sigmoid()
  
  # forward pass
  def forward(self,x,doBN):

    # input (x starts off normalized)
    x = F.relu( self.input(x) )

    if doBN:
      # hidden layer 1
      x = self.bnorm1(x) # batchnorm
      x = self.fc1(x)    # weighted combination
      x = F.relu(x)      # activation function
      # hidden layer 2
      x = self.bnorm2(x) # batchnorm
      x = self.fc2(x)    # weighted combination
      x = F.relu(x)      # activation function
    else:
      # hidden layer 1
      x = F.relu( self.fc1(x) )
      # hidden layer 2
      x = F.relu( self.fc2(x) )

    # output layer
    return self.output(x)


def createANNmodel(learningRate):

  # model architecture
  model = Model()

  # loss function
  lossfun = nn.BCELoss() # but better to use BCEWithLogitsLoss

  # optimizer
  optimizer = torch.optim.SGD(model.parameters(),lr=learningRate)

  # model output
  return model, lossfun, optimizer
```

```python
# a function that trains the model

# a fixed parameter
numepochs = 1000

def trainTheModel(model):

  # initialize losses
  losses = []
  accuracy = []

  # loop over epochs
  for epochi in range(numepochs):
    model.train()
    batchAcc = []
    batchLoss = []

    for X, y in train_loader:
      # forward pass
      yHat = model(X, True)

      # compute loss
      loss = lossfun(yHat,y)

      # backprop
      optimizer.zero_grad()
      loss.backward()
      optimizer.step()

      batchLoss.append(loss.item())
      batchAcc.append(100*torch.mean(((yHat>.5) == y).float()))
  
    losses.append(np.mean(batchLoss)) 
    accuracy.append(np.mean(batchAcc)) 
  
  # # final forward pass
  # predictions = model(data)
    
  # # compute the predictions and report accuracy
  # # NOTE: Wasn't this ">0" previously?!?!
  # totalacc = 100*torch.mean(((predictions>.5) == labels).float())
  
  return losses, accuracy
```

# Test the new code by running it once

```python
# create everything
ANNclassify,lossfun,optimizer = createANNmodel(.01)

# run it
losses, accuracy = trainTheModel(ANNclassify)
```

```python
# show the losses
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))

ax1.plot(losses)
ax1.set_xlabel('Epoch')
ax1.set_ylabel('Loss')
ax2.plot(accuracy)
ax2.set_xlabel('Epoch')
ax2.set_ylabel('Accuracy')
plt.show()
```

# Now for the real test (varying learning rates)

```python
# learning rates
learningrates = np.linspace(.001,.1,50)

# initialize
accByLR = []
allLosses = np.zeros((len(learningrates),numepochs))


# the loop
for i,lr in enumerate(learningrates):
  
  # create and run the model
  ANNclassify,lossfun,optimizer = createANNmodel(lr)
  losses, accuracy = trainTheModel(ANNclassify)

  # store the results
  accByLR.append(accuracy[-1])
  allLosses[i,:] = losses
```

```python
# plot the results
fig,ax = plt.subplots(1,2,figsize=(16,4))

ax[0].plot(learningrates,accByLR,'s-')
ax[0].set_xlabel('Learning rate')
ax[0].set_ylabel('Accuracy')
ax[0].set_title('Accuracy by learning rate')

ax[1].plot(allLosses.T)
ax[1].set_title('Losses by learning rate')
ax[1].set_xlabel('Epoch number')
ax[1].set_ylabel('Loss')
plt.show()
```
